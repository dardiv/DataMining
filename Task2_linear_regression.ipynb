### 1. Dataset loading

```python
from sklearn import datasets

v = 5 #вариант 5

rnd = random.randint(1,20)
print(rnd)


data, target, coef = datasets.make_regression(n_samples = v*10, n_features=v,n_informative =round(v/5),
                                             noise = rnd,coef = True, random_state = 2)
```

### 2. Feature plots

```python
pyplot.figure(figsize(20, 20))

plot_number = 0

for plot_number in range(v):
    pyplot.subplot(4, 4, plot_number + 1)
    pyplot.scatter(list(map(lambda x : x[plot_number], data)), target, color = 'b')
    pyplot.title('feature ' + str(plot_number))
    pyplot.xlabel(str(plot_number))
    pyplot.ylabel('target')
    plot_number += 1
```

### 3. Linear regression model learning

```python
from sklearn import linear_model

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(data, target)
predictions = linear_regressor.predict(data)
```

###  4. Linear regression function

```python
fn_learned = ''
fn_true = ''

for i in range(len(linear_regressor.coef_)):
    fn_learned += str(round(linear_regressor.coef_[i], 3)) + '*x' + str(i + 1) + ' + '
    
fn_learned += str(round(linear_regressor.intercept_, 3))

print ('Learned regression model')
print('y = ' + fn_learned)
```

```python
for i in range(len(coef)):
    fn_true += str(round(coef[i],3)) + '*x' + str(i + 1)+ ' + '

fn_true = fn_true[:len(fn_true)-2]

print ('True regression model')
print('y = ' + fn_true)
```

### 5. Basic estimators

```python
from sklearn import metrics

mea = metrics.mean_absolute_error(target, predictions)
print ('MAE = ', mea)
msa = metrics.mean_squared_error(target, predictions)
print ('MSE = ', msa)
```
